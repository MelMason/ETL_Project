# BTC Cleanup

```python
import pandas as pd
```

```python
# Create a reference the CSV file desired
csv_path = "btc.csv"

# Read the CSV into a Pandas DataFrame
btc_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
btc_df.head()
```

```python

# # Filter the data so that only date for the last 2 years are in a DataFrame
# btc_dates_df = clean_btc_df.loc[(clean_btc_df['date'] >= "7/1/2017")]
# organized_btc_dates = btc_dates_df[["date", "marketcap_usd", "price_usd"]]
# organized_btc_dates.head()
```

```python
# Drop NaN Values
clean_btc_df = btc_df.dropna(how="any")
clean_btc_df.head()
```

```python
# reset index
clean_btc_df.reset_index()
```

```python
# export to csv 
clean_btc_df.to_csv("clean_btc_df.csv")
```

```python
# Filter the data so that only date for the last 2 years are in a DataFrame
clean_btc_df = clean_btc_df.drop(clean_btc_df.index[0:1525])
```

```python
# select only rows needed for analysis
organized_btc_df = clean_btc_df[["date", "marketcap_usd", "price_usd"]]
organized_btc_df.head()
```

```python
# add number of coins column
organized_btc_number = organized_btc_df['marketcap_usd'].div(organized_btc_df['price_usd'])
organized_btc_df['number_of_coins'] = organized_btc_number
organized_btc_df.head()


# # Rename headers
# organized_btc_dates = organized_btc_dates.rename(
#     columns={"date": "Date", "marketcap_usd": "Market Cap (USD)", "price_usd": "Price (USD)"})
# organized_btc_dates.head()
```

```python
# Rename headers
organized_btc_df = organized_btc_df.rename(
    columns={"date": "Date", "marketcap_usd": "Market Cap (USD)", "price_usd": "Price (USD)", "number_of_coins": "Number of Coins"})
organized_btc_df.head()
```

```python
count = organized_btc_df["Date"].count()
print(count)
```

```python
organized_btc_df.to_csv("organized_btc_df.csv")
```

# ADA Cleanup

```python
# Create a reference the CSV file desired
csv_path_2 = "ada.csv"

# Read the CSV into a Pandas DataFrame
ada_df = pd.read_csv(csv_path_2)

# Print the first five rows of data to the screen
ada_df.head()
```

```python
# Drap NaN values
clean_ada_df = ada_df.dropna(how="any")
```

```python
# Filter the data so that only date for the last 2 years are in a DataFrame
ada_dates_df = clean_ada_df.loc[clean_ada_df["date"] >= "7/13/17"]
organized_ada_dates = ada_dates_df[["date", "marketcap_usd", "price_usd"]]
organized_ada_dates.head()
```

```python
# Rename headers
organized_ada_dates = organized_ada_dates.rename(
    columns={"date": "Date", "marketcap_usd": "Market Cap (USD)", "price_usd": "Price (USD)"})
organized_ada_dates.head()
```

```python
count = organized_ada_dates["Date"].count()
print(count)
```

```python
organized_ada_dates.to_csv("organized_ada_dates.csv")
```
